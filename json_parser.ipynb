```python
import pandas as pd
import json
import os
```

```python
def df_per_subject(json_path):
    with open(json_path, 'r') as f:
        example_subject = json.load(f)
    
    #stage 1 material at index 3+1:8
    #stage 2 stability at index 8+1:19
    #stage 3 expense at index 19+1:30
    #stage 4 aesthetics at index 30+1:-1
    
    stabilty_indices = [0,0]
    aesthetic_indices = [0,-1]
    for t in range(len(example_subject)):
        if 'stimulus' in list(example_subject[t].keys()):
            if '<h2>Stage 2' in example_subject[t]['stimulus'][0]:
                stabilty_indices[0] = t+1
            elif '<h2>Stage 3' in example_subject[t]['stimulus'][0]:
                stabilty_indices[1] = t
            elif '<h2>Stage 4' in example_subject[t]['stimulus'][0]:
                aesthetic_indices[0] = t+1
    
    #IPE stability, stability score 
    stability_results = []
    subjid = 'data/experiment3_0dqc41j4zf.json'
    trial_type = 'stability'
    for pair_result in example_subject[stabilty_indices[0]: stabilty_indices[1]]:
        print(pair_result)
        resp = 1+(pair_result['response']/60)*6 #changed from 1-7
        left_video = pair_result['stimulus_left_video']
        right_video = pair_result['stimulus_right_video']

        grouping_no = pair_result['grouping_no']
        stability_results.append([subjid, grouping_no, trial_type, left_video, right_video, resp])
    
    
    #IPE stability, aesthetic score,
    aesthetic_results = []
    subjid = 'data/experiment3_0dqc41j4zf.json'
    trial_type = 'aesthetics'
    for pair_result in example_subject[aesthetic_indices[0]: aesthetic_indices[1]]:
        resp = 1+(pair_result['response']/60)*6 #changed from 1-7
        left_video = pair_result['stimulus_left_video']
        right_video = pair_result['stimulus_right_video']

        grouping_no = pair_result['grouping_no']
        aesthetic_results.append([subjid, grouping_no, trial_type, left_video, right_video, resp])
    
    df = pd.DataFrame(stability_results+aesthetic_results,  columns =  ["subjectID",'grouping_no', 'trial_type', "left_video", 'right_video', "response"])
    return df
```

```python
df_list = []
for path in os.listdir('data'):
    df_list.append(df_per_subject('data/'+path))
```

```python
total = pd.concat(df_list,ignore_index=True)
```
